```python
#Importing Pandas
import pandas as pd
#importing numpy
import numpy as np
```

```python
#Importing data into dataframes
companies = pd.read_csv('companies.txt', sep= "\t", encoding = "ISO-8859-1")
rounds2 = pd.read_csv('rounds2.csv', encoding = "ISO-8859-1")
mapping = pd.read_csv('mapping.csv', encoding = "ISO-8859-1")
```

```python
#Check the columns and first 5 rows of companies file
companies.head()
```

```python
#Check the rounds head
rounds2.head()
```

```python
#To check the shape of both the datasets
print(rounds2.shape, "\n")
print(companies.shape)
```

```python
#Converting company_permalink values into lower case, to compare both datasets companies
companies['permalink'] = companies['permalink'].str.lower()
rounds2['company_permalink'] = rounds2['company_permalink'].str.lower()
```

```python
#No. of unique companies in round2 dataset
unique_rounds2 = rounds2['company_permalink'].unique()
print("No. of unique companies in rounds2 dataset are {}".format(len(unique_rounds2)))
```

```python
#No. of unique companies in companies dataset
unique_companies = companies['permalink'].unique()
print("No. of unique companies in companies dataset are {}".format(len(unique_companies)))
```

```python
#Company dataset description of cunt and unique values
comp_desc = companies.describe()
comp_desc
```

```python
#Finding the unique key in companies dataset
filter = next(index for index, value in enumerate(comp_desc.loc['unique'])if value == len(companies.index))
unique_key = comp_desc.loc['unique'].iloc[[filter]]
print("Name of the unique key which can be used in companies dataset is:",unique_key)
```

```python
#To verify if there is any company/companies in the rounds2 file not present in companies file
comp_present = rounds2.loc[~rounds2['company_permalink'].isin(companies['permalink']),:]
comp_present
```

```python
print("Are there any companies in rounds2 dataset which is not present in companies dataset :\n")
if len(comp_present.index) == 0:
    print("N")
else:
    print("Y,", "There are", len(comp_present.index), "companies.")
```

```python
# Clean the encoding part from companies name in rounds2
rounds2['company_permalink'] = rounds2.company_permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
rounds2.loc[~rounds2['company_permalink'].isin(companies['permalink']),:]
```

```python
# Clean the encoding part from companies name in companies.
companies['permalink'] = companies.permalink.str.encode('utf-8').str.decode('ascii', 'ignore')
companies.loc[~companies['permalink'].isin(rounds2['company_permalink']),:]
```

```python
#Merging the two dataframes so all columns will be added from companies to rounds2 dataframe
master_frame = pd.merge(rounds2, companies, how = 'inner',right_on= 'permalink', left_on = 'company_permalink' )
master_frame
```

```python
#Checking the null values in the data frame
master_frame.isnull().sum()
```

```python
# null percentages in data frame
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#As we are doin analysis based on the funding we can't ignore null values in raised_amount_usd column,so removing rows which contains null values in raised amount
master_frame = master_frame[~np.isnan(master_frame['raised_amount_usd'])]
# checking null percentages in data frame ater droping null value rows from raised_amount_usd 
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#Dropping unnecessary columns
master_frame = master_frame.drop('funding_round_code', axis=1)
master_frame = master_frame.drop('homepage_url', axis=1)
master_frame = master_frame.drop('founded_at', axis=1)
master_frame.head()
```

```python
# More refining of master_frame by removing huge no. of null values from country_code and category_list
master_frame = master_frame[~master_frame.country_code.isnull()]
master_frame = master_frame[~master_frame.category_list.isnull()]
round(100*(master_frame.isnull().sum()/len(master_frame.index)), 2)
```

```python
#Calculating the average funding amount for venture, angel, seed, and private equity.
average_investments = (master_frame.groupby('funding_round_type').raised_amount_usd.mean())/10**6
print("Average investment for Venture is : ",average_investments["venture"])
print("\nAverage investment for Angel is : ",average_investments["angel"])
print("\nAverage investment for Seed is : ",average_investments["seed"])
print("\nAverage investment for Private equity is : ",average_investments["private_equity"])
```

```python
#Venture type of investment is the most suitable type of investment for Spark funds as it lies between 5 to 15 million USD
venture_frame = master_frame[master_frame["funding_round_type"]== "venture"]
venture_frame.head()
```

```python
#Finding top 9 countries which have recieved the highest total funding
top9_countries = venture_frame.groupby('country_code')
top9_countries = (top9_countries['raised_amount_usd'].sum().sort_values(ascending=False))/10**6
top9_countries.head(9)
```

```python
top9 = venture_frame[venture_frame['country_code'].isin(["USA","CHN","GBR","IND","CAN","FRA","ISR","DEU","JPN"])]
top9
```

```python
#splitting the category and creating new column with only primary category.
primary_sector = master_frame["category_list"].str.split("|", n =1 , expand = True) 
master_frame['primary sector']=primary_sector[0]
#Printing dataframe with new column
master_frame.head()
```

```python
#Mapping category list with main_sector.
mappingnew=mapping
col_names = ['Automotive & Sports','Cleantech / Semiconductors','Entertainment','Health','Manufacturing','News, Search and Messaging','Others','Social, Finance, Analytics, Advertising']
def updatemapping(row):
    column = ''
    for col in col_names:
        if row[col] == 1:
            column = col
    return column    
        
mappingnew['main_sector'] = mappingnew.apply(updatemapping,axis=1)
mappingnew.drop(mappingnew.columns[[1,2,3,4,5,6,7,8,9]], axis=1, inplace=True)
mappingnew.head()
```

```python
#Creating new master rame by joining master frame and mapping data frame.
Newmaster_frame = pd.merge(master_frame,mappingnew, how='inner', on = 'category_list')
Newmaster_frame.head()
```

```python
#Creating dataframe for USA with invest ment between 5 to 15 million ,Total invest ment in each sector and total number of investments in each sector
USA_DF = Newmaster_frame[Newmaster_frame.country_code == 'USA']
BolValues=USA_DF['raised_amount_usd'].between(5000000.0, 15000000.0, inclusive=True)
USA_DF=USA_DF[BolValues.values]
total_Investments=USA_DF['raised_amount_usd'].sum()/10**6
print("Total investments is:",total_Investments)
total_number_of_investments=USA_DF['raised_amount_usd'].count()
print("Total number of  investments is:",total_number_of_investments)
groupdf=(USA_DF.groupby(USA_DF['main_sector']).sum().sort_values(by=['raised_amount_usd'],ascending=False))/10**6
groupdf1=USA_DF.groupby(USA_DF['main_sector']).count()
groupdf['Number_of_invetments_per_sector'] = groupdf1['raised_amount_usd'].astype(int)
groupdf.head()
groupdf.rename(columns={'raised_amount_usd':'Total_investment_per_sector(in million)'}, inplace=True)
USA_DF = pd.merge(USA_DF,groupdf, how='inner', on = 'main_sector')
groupdf.head()
USA_DF=USA_DF.sort_values(by=['raised_amount_usd','Number_of_invetments_per_sector'],ascending=False)
pd.set_option('display.max_rows', 100)  # or 1000
USA_DF
```

```python
#Creating dataframe for GBR with invest ment between 5 to 15 million ,Total invest ment in each sector and total number of investments in each sector
GBR_DF = Newmaster_frame[Newmaster_frame.country_code == 'GBR']
BolValues=GBR_DF['raised_amount_usd'].between(5000000.0, 15000000.0, inclusive=True)
GBR_DF=GBR_DF[BolValues.values]
total_Investments=GBR_DF['raised_amount_usd'].sum()/10**6
print("Total investments is:",total_Investments)
total_number_of_investments=GBR_DF['raised_amount_usd'].count()
print("Total number of  investments is:",total_number_of_investments)
groupdf=(GBR_DF.groupby(GBR_DF['main_sector']).sum().sort_values(by=['raised_amount_usd'],ascending=False))/10**6
groupdf1=GBR_DF.groupby(GBR_DF['main_sector']).count()
groupdf['Number_of_invetments_per_sector'] = groupdf1['raised_amount_usd'].astype(int)
groupdf.head()
groupdf.rename(columns={'raised_amount_usd':'Total_investment_per_sector(in million)'}, inplace=True)
GBR_DF = pd.merge(GBR_DF,groupdf, how='inner', on = 'main_sector')
groupdf.head()
GBR_DF=GBR_DF.sort_values(by=['raised_amount_usd','Number_of_invetments_per_sector'],ascending=False)
GBR_DF
```
